## Задание 3

```python
np.random.seed(4)


lam = 1.5
pois_dist = np.random.poisson(1.5, 10001)
plt.hist(pois_dist, max(pois_dist), density=1, facecolor='g', alpha=0.75)

x = np.arange(min(pois_dist), max(pois_dist))
y = [((lam**i * np.exp(-lam)) / np.math.factorial(i)) for i in x]
plt.plot(x, y, '--')
```
