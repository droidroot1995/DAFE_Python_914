```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

### Реализовать численное решение уравнения теплопроводности с помощью метода конечных разностей для представленного выше случая, построить график распределения температуры по толщине пластины.

```python
T0 = 100
Tl = 1000
Tr = 400
L = 0.1
Lambda = 50
C = 460
p = 7800
Te = 300

dx = 0.01
dt = 0.01
Nx = round(L/dx)
Nt = round(Te/dt)

T = np.zeros((Nx, Nt), dtype = np.float64)
T[0][0] = Tl
T[Nx-1][0] = Tr
for i in range(1,Nx-1):
    T[i][0] = T0


a = Lambda * dt / (p * C * (dx**2))
for j in range(1, Nt):
    T[0][j] = Tl
    T[Nx-1][j] = Tr
    for i in range(1,N-1):
        T[i][j] = T[i][j-1] + a*(T[i+1][j-1] - 2*T[i][j-1] + T[i-1][j-1])
        
x = np.linspace(0, L, Nx)
Te = [T[i][Nt-1] for i in range(Nx)]

fig, ax = plt.subplots(figsize=[7, 5])
ax.plot(x, Te)
ax.set_title("Распределение температуры по пластине T=T(x)")
plt.show()
```
