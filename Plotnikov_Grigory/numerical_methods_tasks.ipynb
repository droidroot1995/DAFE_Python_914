### Task1: –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ä–∞—Å—á–µ—Ç –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –≤ —Ç–æ—á–∫–µ –∏–∑ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞  [ùë•0,ùë•1] , –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –µ—ë –≥—Ä–∞—Ñ–∏–∫

```python
import numpy as np
import matplotlib.pyplot as plt


def f1(x):
    return (np.exp(x) - 1/2)**1/2

def f2(x):
    return np.exp(x)**(-3 * x) * x**3

dx = 0.1
x = np.arange(0, 10, dx)

df1 = np.zeros(len(x))
df2 = np.zeros(len(x))

for i in range(len(x) -  1):
    df1[i] = (f1(x[i + 1]) - f1(x[i - 1])) / (2 * dx)
    df2[i] = (f2(x[i + 1]) - f2(x[i - 1])) / (2 * dx)

fig, axes = plt.subplots(ncols=2, nrows=1, figsize=[15, 5])
axes[0].plot(x, df1)
axes[0].set_title('derivative of y=(e^x - 1/2)^1/2')
axes[1].plot(x, df2)
axes[1].set_title('derivative of y=e^(-3x) * x^3')
plt.show()
```

### Task2: –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏–Ω—Ç–µ–≥—Ä–∞–ª–∞ –º–µ—Ç–æ–¥–æ–º –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–æ–≤

```python
def integrate1(f, a, b, delta=0.001):
    return dx * sum(f(np.arange(a, b, delta)))
```

### Task3: –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏–Ω—Ç–µ–≥—Ä–∞–ª–∞ –º–µ—Ç–æ–¥–æ–º —Ç—Ä–∞–ø–µ—Ü–∏–∏

```python
def integrate2(f, a, b, delta=0.001):
    x = np.arange(a, b, delta)
    return dx * 0.5 * sum(f(x[i]) + f(x[i + 1]) for i in range(len(x) - 1))
```

### Task4: –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏–Ω—Ç–µ–≥—Ä–∞–ª–∞ –º–µ—Ç–æ–¥–æ–º –°–∏–º–ø—Å–æ–Ω–∞

```python
def integrate3(f, a, b, delta=0.001):
    x = np.arange(a, b, delta)
    np.array
    return dx * 1/3 * (f(x[0]) + f(x[-1]) + 2 * sum(f(x[i]) for i in range(2, len(x) - 1, 2))
                                          + 4 * sum(f(x[i]) for i in range(1, len(x) - 1, 2)))
```

### Task5: –í—ã—á–∏—Å–ª–∏—Ç—å –∏–Ω—Ç–µ–≥—Ä–∞–ª –∑–∞–¥–∞–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ –≤—Å–µ–º–∏ —Ç—Ä–µ–º—è —Å–ø–æ—Å–æ–±–∞–º–∏, –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –µ—ë –≥—Ä–∞—Ñ–∏–∫ –∏ —Å—Ä–∞–≤–Ω–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤—ã—á–∏—Å–ª–µ–Ω–∏—è.

```python
def f(x):
    return np.sin(x)

fig, ax = plt.subplots(figsize=[7, 5])
args = [f, 0, 4*np.pi, 0.001]
x = np.arange(args[1], args[2], args[3])
ax.plot(x, f(x))
plt.show()

print('integration results:')
print(f'1: {integrate1(*args)}\n2: {integrate2(*args)}\n3: {integrate3(*args)}')
```
