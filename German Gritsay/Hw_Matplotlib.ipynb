#### Задание: нарисовать квадрат с вершинами в точках (1, 1), (1, -1), (-1, -1), (-1, 1)

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
rect = matplotlib.patches.Rectangle((-1, -1), width = 2, height = 2)
ax.add_patch(rect)
plt.axis([-2, 2, -1.2, 1.2])
plt.show()
```

#### Задание: построить точечный график y = sin(x)

```python
import numpy as np

x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.scatter(x, np.sin(x))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Задание: построить гистограмму распределения Пуассона$$p(x) = \frac{e^{-\lambda}\lambda^x}{x!}$$, при$$\lambda = 1.5; x = 0, 1, ..., 10000$$, с использованием numpy.random

```python
import math

lmd = 10

poi = np.random.poisson(lmd, 10000)
maxx = max(poi)
plt.hist(poi, maxx, density=1, facecolor='b', alpha=0.75)
plt.text(20, .025, r'$\lambda=1.5$')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Poisson Distribution')
plt.show()
```
