```python
import matplotlib.pyplot as plt
import numpy as np
```

f(x) = sin(x); [0, 2*pi]

```python
def f(x):
    return np.sin(x)
a = 0
b = 2*np.pi
n = 20
```

```python
x = np.linspace(a, b, n)
fx = f(x)
h = (b-a)/n

dfx = np.zeros(n, float)
dfx[0] = (fx[1]-fx[0])/h
for i in range(1,n-1):
    dfx[i] = (fx[i+1]-fx[i-1])/2/h
dfx[n-1] = (fx[i]-fx[i-1])/h

plt.plot(x, dfx, label="(sinx)'")
plt.plot(x, np.cos(x) , label='cosx')
plt.xlabel('x label')
plt.ylabel('y label') 
plt.show()
```

```python
# средние прямоугольники
x = np.linspace(a, b, n)
h = (b-a)/n
integral = np.zeros(n, float) 
for i in range(1,n):
    integral[i] = integral[i-1] + f(x[i]+h/2)*h
    
plt.plot(x, integral, label='integral(sinx)')
plt.plot(x, -np.cos(x)+1, label="-cos(x)+1")
plt.xlabel('x label')
plt.ylabel('y label') 
plt.show()
```

```python
# трапеции
x = np.linspace(a, b, n)
integral = np.zeros(n, float) 
h = (b-a)/n
for i in range(1,n):
    integral[i] = integral[i-1] + (f(x[i]-h)+f(x[i]))*h/2
    
plt.plot(x, integral, label='integral(sinx)')
plt.plot(x, -np.cos(x)+1, label="-cos(x)+1")
plt.xlabel('x label')
plt.ylabel('y label') 
plt.show()
```

```python
# Симпсон
x = np.linspace(a, b, n)
h = (b-a)/n
integral = np.zeros(n, float)
for i in range(1,n):
    integral[i] = integral[i-1] + (f(x[i]-h)+f(x[i])+4*f(x[i]-h/2))*h/6

plt.plot(x, integral, label='integral(sinx)')
plt.plot(x, -np.cos(x)+1, label="-cos(x)+1")
plt.xlabel('x label')
plt.ylabel('y label') 
plt.show()
```

```python
a = 0
b = 10
h = 0.1
n = int((b-a)/0.1)
def f(x,y):
    return x*np.sqrt(y)

def fi(x):
    return (x**2 + 4)**2 / 16
y0 = float(1.)
```

```python
#Эйлер
x = np.linspace(a, b, n)
y = np.zeros(n, float)
y[0] = y0
for i in range(1,n):
    y[i] = y[i-1] + h*f(x[i-1], y[i-1])   

plt.plot(x, y , label='1')
plt.plot(x, fi(x), label="2")
plt.xlabel('x label')
plt.ylabel('y label') 
plt.show()
```

```python
# Рунге—Кутт
x = np.linspace(a, b, n)
y = np.zeros(n, float)
y[0] = y0
for i in range(1,n):
    k1 = f(x[i-1], y[i-1])
    k2 = f(x[i-1] + h/2, y[i-1] + h*k1/2)
    k3 = f(x[i-1] + h/2, y[i-1] + h*k2/2)
    k4 = f(x[i-1] + h, y[i-1] + h*k3)
    y[i] = y[i-1] + h*(k1+2*k2+2*k3+k4)/6   

plt.plot(x, y , label='1')
plt.plot(x, fi(x), label="2")
plt.xlabel('x label')
plt.ylabel('y label') 
plt.show()
```
