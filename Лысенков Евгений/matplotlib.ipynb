```python
import matplotlib.pyplot as plt
import numpy as np
x = x = np.linspace(0, 2*np.pi, 50)
plt.axis([0, 7, -1.2, 1.2])
plt.scatter(x, np.sin(x))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

lam = 10

poi = np.random.poisson(lam, 100000)
maxx = max(poi)
plt.hist(poi, maxx, density=1, facecolor='g', alpha=0.75)

x = range(maxx)
y = []
for i in range(maxx):
    y.append((lam**i * np.exp(-lam)) / factorial(i))


plt.plot(x, y, '--')
plt.text(0, .12, r'$\lam=1.5')
plt.grid(True)
plt.show()
```
