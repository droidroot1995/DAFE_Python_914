# Отрисовываем квадрат с нужными вершинами


```python
fig, ax = plt.subplots()
rectangle = matplotlib.patches.Rectangle((-1,-1),width = 2, height = 2, color = "black", fill=True) 
ax.add_patch(rectangle)
plt.axis([-2, 2, -1.2, 1.2])
plt.show()
```

# Рисуем график sin(x) по точкам


```python
from math import sin as sin
import matplotlib.lines as mlines
ax = plt.gca()
x = [i for i in range(-10,10)]
y = []
for i in x:
    y.append(sin(i))
plt.axis([-10, 10, -10, 10])
plt.scatter(x, y) 
plt.xlabel('x')
plt.ylabel('y')
l = mlines.Line2D([-10,10], [0,0], color = 'black')
r = mlines.Line2D([0,0], [-10,10], color = 'black')
ax.add_line(l)
ax.add_line(r)
plt.show()
```

# Распределение Пуассона

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

lam = 10

poi = np.random.poisson(lam, 100000)
maxx = max(poi)
plt.hist(poi, maxx, density=1, facecolor='g', alpha=0.75)

x = range(maxx)
y = []
for i in range(maxx):
    y.append((lam**i * np.exp(-lam)) / factorial(i))


plt.plot(x, y, '--')
plt.text(0, .12, r'$\lam=1.5')
plt.grid(True)
plt.show()
```
