```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

# Квадрат

```python
fig, ax = plt.subplots()
square = matplotlib.patches.Rectangle((-1, -1), 2, 2, fill=True)
ax.add_patch(square)
plt.axis([-2, 2, -1.2, 1.2])
```

# Точечный синус

```python
x = np.linspace(0, 2 * np.pi, 50)
plt.scatter(x, np.sin(x))
plt.xlabel('x')
plt.ylabel('y')
```

# Гистограмма

```python
lam = 10

poi = np.random.poisson(lam, 100000)
maxx = max(poi)
plt.hist(poi, maxx, density=1, facecolor='blue', alpha=0.75)

x = range(maxx)
y = []
for i in range(maxx):
    y.append((lam**i * np.exp(-lam)) / math.factorial(i))

plt.plot(x, y)
plt.grid(True)
```
