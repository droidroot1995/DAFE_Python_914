```python
import matplotlib.pyplot as plt
import numpy as np
```

### f(x) = cos(x); [0, 2*pi] ###

```python
def f(x):
    return np.cos(x)
a = 0
b = 2 * np.pi
n = 20
```

# Двусторонняя разность #

```python
x = np.linspace(a, b, n)
fx = f(x)
step = (b-a) / n

dfx = np.zeros(n, float)
dfx[0] = (fx[1]-fx[0]) / 2 * step
for i in range(1, n-1):
    dfx[i] = (fx[i+1]-fx[i-1]) / (2 * step)
dfx[n-1] = (fx[i]-fx[i-1]) / (2 * step)

plt.plot(x, dfx, label="(cosx)'")
plt.plot(x, -np.sin(x) , label='-sinx')
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
```

# Средние прямоугольники

```python
x = np.linspace(a, b, n)
step = (b-a) / n

integ = np.zeros(n, float) 
for i in range(1,n):
    integ[i] = integ[i-1] + f(x[i]+step/2) * step
    
plt.plot(x, integ, label='integral(cosx)')
plt.plot(x, np.sin(x) - 0.35, label="sin(x) - 0.35")
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
```

# Трапеции #

```python
x = np.linspace(a, b, n)
integ = np.zeros(n, float) 
step = (b-a) / n
for i in range(1,n):
    integ[i] = integ[i-1] + (f(x[i]-step)+f(x[i])) * (step / 2)
    
plt.plot(x, integ, label='integral(cosx)')
plt.plot(x, np.sin(x), label="sinx")
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
```

# Метод Симпсона #

```python
x = np.linspace(a, b, n)
integ = np.zeros(n, float) 
step = (b-a) / n
for i in range(1,n):
    integ[i] = integ[i-1] + (f(x[i]-step) + f(x[i]) + 4 * f(x[i]-step/2)) * (step / 6)

plt.plot(x, integ, label='integral(cosx)')
plt.plot(x, np.sin(x), label="sinx")
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
```

```python
a = 0
b = 10
h = 0.1
n = int((b-a)/0.1)
def f(x,y):
    return x * np.sqrt(y)

def fi(x):
    return (x**2 + 4)**2 / 16
y0 = 1
```

# Эйлер

```python
x = np.linspace(a, b, n)
y = np.zeros(n, float)
y[0] = y0
for i in range(1,n):
    y[i] = y[i-1] + h * f(x[i-1], y[i-1])   

plt.plot(x, y , label='1')
plt.plot(x, fi(x), label="2")
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
```

# Рунге-Кутт

```python
x = np.linspace(a, b, n)
y = np.zeros(n, float)
y[0] = y0
for i in range(1,n):
    k1 = f(x[i-1], y[i-1])
    k2 = f(x[i-1] + h/2, y[i-1] + h * k1/2)
    k3 = f(x[i-1] + h/2, y[i-1] + h * k2/2)
    k4 = f(x[i-1] + h, y[i-1] + h * k3)
    y[i] = y[i-1] + h * (k1 + 2 * k2 + 2 * k3 + k4)/6   

plt.plot(x, y , label='1')
plt.plot(x, fi(x), label="2")
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
```
