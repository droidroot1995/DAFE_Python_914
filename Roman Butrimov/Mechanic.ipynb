# Absolute elastic

```python
from IPython.display import HTML
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

fig, ax = plt.subplots()

v0 = 0
h = 10
g = 9.8
y0 = h
dt = 0.1
T = (2*h/g)**0.5
x = np.arange(0, 5*T, 0.01)
ax.set_ylim(0, h+5)
ax.set_xlim(-1, 1)

def init():
    line = ax.scatter([0], [0])
    return line


def animate(i):
    ax.clear()
    ax.set_ylim(0, h+5)
    ax.set_xlim(-1, 1)
    A = i // T
    i = i - T*A
    if (A%2 == 0):
        line = ax.scatter([0], [h - g*(i**2)/2])
    else:
        line = ax.scatter([0], [g*(i**2)/2])
    return line


ani = animation.FuncAnimation(
    fig, animate, init_func=init, interval=2) #, blit=True, save_count=50)

HTML(ani.to_jshtml())


```

```python
fig, ax = plt.subplots()

v0 = 0
m = 1
h = 10
g = 9.8
y0 = h
dt = 0.1
T = (2*h/g)**0.5
Vmax = g*T
x = np.arange(0, 5*T, 0.01)

time = list()
for a in x:
    A = a // T
    a = a - T*A
    if (A%2 == 0):
        time.append(g*a)
    else:
        time.append(Vmax-g*a)

plt.plot(x, time)
plt.show()
```
