```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def diff(f, a, b, x0):
    
    delta = min(0.001, (b - a) / 1000)
    
    return (f(min(x0 + delta, b)) - f(max(x0 - delta, a))) / (min(x0 + delta, b) - max(x0 - delta, a))
```

```python
def diff_segment(f, a, b, num):
    
    segment = np.linspace(a, b, num)
    
    return [diff(f, a, b, i) for i in segment]
```

```python
fig, axes = plt.subplots(4, 1, figsize=(9, 32))
titles = ["$sin x$", "$cos x$", "$x ^ 2$", "$x ^ 3$"]

num = 100

for i, data in enumerate([(np.sin, 0, 2 * np.pi), 
                          (np.cos, 0, 2 * np.pi), 
                          (lambda x: x ** 2, -1, 4), 
                          (lambda x: x ** 3, -1, 1)]):

    axes[i].grid(linestyle='--')
    axes[i].plot(np.linspace(data[1], data[2], num), diff_segment(data[0], data[1], data[2], num))
    axes[i].set_title(titles[i])
```
