```python
import matplotlib.pyplot as plt
from matplotlib import patches
import numpy as np
```

#### Задание: нарисовать квадрат с вершинами в точках (1, 1), (1, -1), (-1, -1), (-1, 1)

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))

ax.grid(linestyle='--')
rect = patches.Rectangle((-1, -1), 2, 2)
ax.add_patch(rect)
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

ax.set_title('square figure')
plt.show()
```

#### Задание: построить точечный график y = sin(x)

```python
x = np.linspace(-np.pi, np.pi, 100)

plt.scatter(x, np.sin(x), marker='.')
plt.title("$sinx$ scatter graphic")
plt.xlabel('$x$')
plt.ylabel('$sinx$')
plt.show()
```

Задание: построить гистограмму распределения Пуассона$$p(x) = \frac{e^{-\lambda}\lambda^x}{x!}$$, при$$\lambda = 1.5; x = 0, 1, ..., 10000$$, с использованием numpy.random

```python
np.random.seed(2)
```

```python
lam = 1.5

h = np.random.poisson(lam, 100000)
x = np.arange(min(h), max(h) + 1, 1, int)
y = np.exp(-lam) * lam ** x / np.array([np.math.factorial(i) for i in x])
```

```python
fig,axes = plt.subplots()

axes.hist(h, density=True, label='experimental')
axes.plot(x + 0.5, y, marker='.', label='calculated')

plt.legend()
axes.set_title("Poisson distribution")
axes.set_ylabel("Probability")
axes.set_xlabel("x")
plt.show()
```
