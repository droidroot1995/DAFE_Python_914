### Задание: нарисовать квадрат с вершинами в точках (1, 1), (1, -1), (-1, -1), (-1, 1)

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.cm as cm
import matplotlib.gridspec as gridspec
from matplotlib.ticker import NullFormatter
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
from matplotlib.ticker import LinearLocator
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import axes3d

fig, ax = plt.subplots()

rec = matplotlib.patches.Rectangle((-1, -1), 2, 2)
ax.add_patch(rec)
plt.axis([-2, 2, -1.2, 1.2])
plt.show()
```

### Задание: построить точечный график y = sin(x)

```python
x = np.arange(0, 2 * np.pi, 0.5)
y = np.sin(x)

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

### Задание: построить гистограмму распределения Пуассона

$ p(k) = \frac{\lambda ^ {k}}{k!} e^{-\lambda}$ 

$\lambda = 1.5$

$ x \in [0, 10000]$

```python
import scipy.special

lam = 1.5

x = np.random.poisson(lam, 10000)
n, bins, patches = plt.hist(x, np.max(x), density=1, facecolor='g', alpha=0.75)

y = np.power(lam, bins) / scipy.special.factorial(bins)  * np.exp(-lam)

plt.plot(bins, y, '--')
plt.show
```
